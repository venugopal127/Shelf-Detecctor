```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model

# Paths for the dataset
train_dir = "C:/Users/venug/Downloads/INVENtoryyy/train"
val_dir = "C:/Users/venug/Downloads/INVENtoryyy/valid"
test_dir = "C:/Users/venug/Downloads/INVENtoryyy/test"

# Image Data Generators with Augmentation
train_datagen = ImageDataGenerator(
    zoom_range=0.5,
    rescale=1/255,
    horizontal_flip=True,
    rotation_range=40,  # Added rotation for augmentation
    width_shift_range=0.2,  # Added width shift for augmentation
    height_shift_range=0.2,  # Added height shift for augmentation
    shear_range=0.2,  # Added shear range for augmentation
    fill_mode='nearest',  # Fill mode for edge cases
    preprocessing_function=preprocess_input
)

val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train = train_datagen.flow_from_directory(
    directory=train_dir,
    target_size=(256, 256),  # Changed from (128, 128) to (256, 256)
    batch_size=8,
    class_mode='binary'
)

val = val_datagen.flow_from_directory(
    directory=val_dir,
    target_size=(256, 256),  # Changed from (128, 128) to (256, 256)
    batch_size=8,
    class_mode='binary'
)

# Base Model: ResNet50
base_model = ResNet50(input_shape=(256, 256, 3), include_top=False, weights='imagenet')

# Freeze the layers of the base model
for layer in base_model.layers:
    layer.trainable = False

# Adding Custom Layers
x = Flatten()(base_model.output)
x = Dense(1, activation='sigmoid')(x)  # Sigmoid for binary classification
model = Model(base_model.input, x)

# Compile the Model with a learning rate scheduler
model.compile(optimizer=Adam(learning_rate=1e-4), loss='binary_crossentropy', metrics=['accuracy'])

# Callbacks for EarlyStopping, ModelCheckpoint, and Learning Rate Reduction
es = EarlyStopping(monitor='val_accuracy', min_delta=0.01, patience=5, verbose=1, restore_best_weights=True)
mc = ModelCheckpoint(filepath="best_model_shelf.keras", monitor='val_accuracy', save_best_only=True, verbose=1)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6, verbose=1)

# Training the Model
steps_per_epoch = train.samples // train.batch_size
validation_steps = val.samples // val.batch_size

history = model.fit(
    train,
    steps_per_epoch=steps_per_epoch,
    epochs=30,  # Increased the epochs to 30 for better convergence
    verbose=1,
    callbacks=[es, mc, lr_scheduler],
    validation_data=val,
    validation_steps=validation_steps
)

# Load the Best Model for Prediction
model = load_model("best_model_shelf.keras")

# Evaluate the model on the test set
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_generator = test_datagen.flow_from_directory(
    directory=test_dir,
    target_size=(256, 256),
    batch_size=8,
    class_mode='binary'
)

test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy:.2f}")
print(f"Test Loss: {test_loss:.2f}")

# Class Reference Mapping
class_labels = list(train.class_indices.keys())
ref = dict(zip(list(train.class_indices.values()), class_labels))

# Example: Predicting a single image
from tensorflow.keras.preprocessing import image

img_path = 'path_to_image.jpg'  # Change this path to the image you want to predict
img = image.load_img(img_path, target_size=(256, 256))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

prediction = model.predict(img_array)
predicted_class = ref[int(prediction > 0.5)]
print(f"Predicted Class: {predicted_class}")
```

```python
import cv2
from inference_sdk import InferenceHTTPClient
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from plyer import notification
import time  # For cooldown implementation

CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",  
    api_key="VQlbuYKgzgBMPY0UNrgl" 
)

MODEL_ID = "inventario-v2-390/4"

# Main pipeline controller
def shelf_monitor_pipeline(video_source="rtsp://192.168.1.104:8080/h264.sdp"):
    # Start video capture (use RTSP stream URL as the source)
    cap = cv2.VideoCapture(video_source)
    
    if not cap.isOpened():
        print("Error: Cannot connect to RTSP stream.")
        return

    ret, prev_frame = cap.read()
    prev_frame_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    prev_frame_gray = cv2.GaussianBlur(prev_frame_gray, (21, 21), 0)

    last_capture_time = 0  # Time of last captured image (initially set to 0)
    cooldown_period = 20  # Cooldown period in seconds

    while True:
        # Step 2: Motion Detection
        ret, frame = cap.read()
        if not ret:
            print("Error: Cannot read frame from stream.")
            break

        current_time = time.time()
        if detect_motion(prev_frame_gray, frame):
            if current_time - last_capture_time > cooldown_period:
                # Step 3: Capture Image
                image_path = capture_image(frame)
                last_capture_time = current_time  # Update last capture time

                # Step 4: Classify Image
                if check_empty_space_with_api(image_path):

                    notify_shopkeeper()


        prev_frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        prev_frame_gray = cv2.GaussianBlur(prev_frame_gray, (21, 21), 0)

        cv2.imshow("Shelf Monitor Pipeline", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

def detect_motion(prev_frame_gray, current_frame):
    gray = cv2.cvtColor(current_frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (21, 21), 0)
    frame_delta = cv2.absdiff(prev_frame_gray, gray)
    threshold = cv2.threshold(frame_delta, 25, 255, cv2.THRESH_BINARY)[1]
    contours, _ = cv2.findContours(threshold.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    motion_detected = any(cv2.contourArea(contour) > 500 for contour in contours)
    if motion_detected:
        print("Motion detected!")
    return motion_detected


def capture_image(frame):
    image_path = "detected_image.jpg"
    cv2.imwrite(image_path, frame)
    print("Image captured.")
    return image_path

def check_empty_space_with_api(image_path):
    result = CLIENT.infer(image_path, model_id=MODEL_ID)

    if 'predictions' in result and len(result['predictions']) > 0:
        for prediction in result['predictions']:
            
            if prediction['class'].lower() in ["vacio", "empty"]:  
                print("The image shows: Empty Space")
                return True  
        print("The image shows: Occupied Space")
        return False  
    else:
        print("No predictions were made, assuming occupied space.")
        return False


def notify_shopkeeper():
    notification.notify(
        title="Shelf Monitor Alert",
        message="An empty space has been detected on the shelf. Please restock it.",
        app_name="Shelf Monitor"
    )
    print("Desktop notification sent.")


shelf_monitor_pipeline()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load the sales data
df = pd.read_csv("sales_data.csv")

# Filter data for a specific item (e.g., "Item A")
item_name = "Item A"
item_data = df[df["Item"] == item_name]

# Aggregate sales by date
item_data = item_data.groupby("Date").agg({"Quantity Sold": "sum"}).reset_index()

# Rename columns to match Prophet requirements
item_data.rename(columns={"Date": "ds", "Quantity Sold": "y"}, inplace=True)

# Convert date column to datetime format
item_data["ds"] = pd.to_datetime(item_data["ds"])

# Initialize and fit the Prophet model
model = Prophet()
model.fit(item_data)

# Make future predictions
future = model.make_future_dataframe(periods=30)  # Forecast for the next 30 days
forecast = model.predict(future)

# Calculate recommended stock levels
forecast["required_stock"] = forecast["yhat"].apply(lambda x: max(0, int(x)))  # Assuming no negative stock

# Plot the forecast
model.plot(forecast)
plt.title(f"Sales Forecast and Stock Requirement for {item_name}")
plt.xlabel("Date")
plt.ylabel("Quantity Sold")
plt.show()

# Save the forecast with stock requirements to a CSV file
forecast[["ds", "yhat", "yhat_lower", "yhat_upper", "required_stock"]].to_csv(f"{item_name}_forecast_with_stock.csv", index=False)
print(f"Forecast with stock requirements saved as '{item_name}_forecast_with_stock.csv'")
```
